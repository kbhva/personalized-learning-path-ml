```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
%matplotlib inline
```

```python
df = pd.read_csv("../datasets/resources.csv")
df.head()
```

```python
df.info()
df.isnull().sum()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(y="topic", data=df, order=df['topic'].value_counts().index)
plt.title("Number of Resources per Topic")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x="length", y="popularity", hue="difficulty", data=df)
plt.title("Length vs Popularity colored by Difficulty")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df["rating"], bins=20, kde=True)
plt.title("Rating Distribution")
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(x="difficulty", data=df)
plt.title("Difficulty Distribution")
plt.show()
```
